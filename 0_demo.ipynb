## Hardware Generators: Type-Safe Meta-Programming for RTL

모든 하드웨어 기술 언어는 RTL 디자인의 단일 인스턴스 작성을 지원합니다. 치즐도 다르지 않습니다.
사실, 대부분의 Verilog/VHDL 디지털 로직 디자인은 Chisel로 직접 옮겨질 수 있습니다!
Chisel은 우리가 얻을 수 있는 다른 멋진 기능을 제공하지만 Chisel로 전환하는 사용자는 다른 하드웨어 언어와 동일한 수준으로 디자인을 제어할 수 있다는 점을 강조하고 싶습니다.

FIR 필터 스타일로 구현된 3-point moving average의 다음 예를 살펴보겠습니다.

<img src="images/demo_fir_filter.svg" width="512" />

Chisel은 합성 가능한 Verilog와 유사한 기본 프리미티브를 제공하며 그대로 *사용* 가능합니다! 다음 셀을 실행하여 Chisel 모듈을 선언합니다.

```scala
// 3-point moving average implemented in the style of a FIR filter
class MovingAverage3(bitWidth: Int) extends Module {
  val io = IO(new Bundle {
    val in = Input(UInt(bitWidth.W))
    val out = Output(UInt(bitWidth.W))
  })

  val z1 = RegNext(io.in) // Create a register whose input is connected to the argument io.in
  val z2 = RegNext(z1)    // Create a register whose input is connected to the argument z1

  io.out := (io.in * 1.U) + (z1 * 1.U) + (z2 * 1.U) // `1.U` is an unsigned literal with value 1
}
```

`class MovingAverage3`을 정의한 후 이를 인스턴스화하고 구조를 살펴보겠습니다.

```scala
// same 3-point moving average filter as before
visualize(() => new MovingAverage3(8))
```

Chisel 인스턴스의 이 시각화에서 왼쪽의 입력과 z1 및 z2는 금색으로 등록됩니다. 레지스터와 io_in 모두 해당 계수로 곱해진 다음 연속적으로 추가됩니다. `tail` 및 `bits` 요소는 추가가 증가하는 것을 방지하는 데 사용됩니다.


이제 다음과 같이 질문할 수 있습니다. "오, 좋습니다. Chisel의 Verilog에서 작업을 수행할 수 있지만 왜 Chisel을 사용하고 싶습니까?"

질문해주셔서 정말 기쁩니다! Chisel의 진정한 힘은 **인스턴스가 아닌 제너레이터**를 생성하는 능력에서 나옵니다. `MovingAverage3` 모듈 대신 계수 목록으로 매개변수화된 일반 `FIRFilter` 모듈을 만들고 싶었다고 가정해 보겠습니다.


아래에서 일련의 계수로 받아들이도록 `MovingAverage3`을 다시 작성했습니다. 계수의 수가 필터의 크기를 결정합니다.

```scala
// Generalized FIR filter parameterized by the convolution coefficients
class FirFilter(bitWidth: Int, coeffs: Seq[UInt]) extends Module {
  val io = IO(new Bundle {
    val in = Input(UInt(bitWidth.W))
    val out = Output(UInt())
  })
  // Create the serial-in, parallel-out shift register
  val zs = Reg(Vec(coeffs.length, UInt(bitWidth.W)))
  zs(0) := io.in
  for (i <- 1 until coeffs.length) {
    zs(i) := zs(i-1)
  }

  // Do the multiplies
  val products = VecInit.tabulate(coeffs.length)(i => zs(i) * coeffs(i))

  // Sum up the products
  io.out := products.reduce(_ +& _)
}
```

이제 인스턴스화 중에 `coeffs` 매개변수를 변경하여 `FIRFilter` 모듈을 사용하여 무한한 수의 다양한 하드웨어 모듈을 인스턴스화할 수 있습니다! 아래에서 `FIRFiler`의 세 가지 다른 인스턴스를 만듭니다.

```scala
// same 3-point moving average filter as before
visualize(() => new FirFilter(8, Seq(1.U, 1.U, 1.U)))
```

```scala
// 1-cycle delay as a FIR filter
visualize(() => new FirFilter(8, Seq(0.U, 1.U)))
```

```scala
// 5-point FIR filter with a triangle impulse response
visualize(() => new FirFilter(8, Seq(1.U, 2.U, 3.U, 2.U, 1.U)))
```

이 강력한 매개변수화 없이는 각 FIR 필터에 대해 하나씩 더 많은 모듈 정의가 필요합니다. 이상적으로는 생성기가 (1) 구성 가능하고, (2) 강력하고, (3) 생성된 디자인을 세밀하게 제어할 수 있기를 바랍니다.


Chisel의 장점은 언어 자체가 아니라 사용 방법에 있습니다.
생성기 대신 인스턴스를 작성하기로 결정하면 Verilog보다 Chisel의 장점이 더 적습니다.
그러나 시간을 내어 제너레이터를 작성하는 방법을 배우면 Chisel의 힘이 분명해지고 Verilog 작성으로 돌아갈 수 없다는 것을 깨닫게 될 것입니다.
제너레이터 작성을 배우는 것은 어렵지만 이 튜토리얼이 여러분이 더 나은 하드웨어 디자이너, 프로그래머, 사상가가 되는 길을 열어주기를 바랍니다!
